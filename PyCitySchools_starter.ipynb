```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(10)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
SchoolCount = len(school_data_complete.school_name.unique())
TotalStudents = len(school_data_complete.student_name.unique())
TotalBudget = school_data_complete.budget.sum()
AvgMathScore = school_data_complete.math_score.mean().round(2)
AvgReadingScore = school_data_complete.reading_score.mean().round(2)
PassingMath_df = school_data_complete.loc[school_data_complete['math_score'] >= 70]
TotalStudentsPassingMath = len(PassingMath_df)
PercentPassingMath = ((TotalStudentsPassingMath / TotalStudents) *100)
PassingReading_df = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
TotalStudentsPassingReading = len(PassingReading_df)
PercentPassingReading = ((TotalStudentsPassingReading / TotalStudents) *100)

Passing_df = student_data.loc[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)]

TotalStudentsPassing = len(Passing_df)

PercentTotalPassing = ((TotalStudentsPassing/TotalStudents) * 100)


District_Summary_df = pd.DataFrame({'Number of Schools': [SchoolCount], 'Total Students': [TotalStudents], 'Total Budget': [TotalBudget],
                                   'Average Math Score': [AvgMathScore], 'Percent of Students Passing Math':[PercentPassingMath],
                                   'Average Reading Score': [AvgReadingScore],'Percent of Students Passing Reading':[PercentPassingReading],
                                   'Percent of Students Passing Reading and Math': [PercentTotalPassing]})
District_Summary_df
```
